USA Housing


```python
#DOC CSV
import pandas as pd
url = 'USA_Housing.csv'
data = pd.read_csv(url)
data.head
#VE BIEU DO
import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()
```

```python
correlation = data.corr()
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot=True)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Avg. Area Income']),y=data['Price'])
print('Gia tri Intercept la: ',model.intercept_)
print('Gia tri Coeffient la: ',model.coef_)
model.score(X=pd.DataFrame(data['Avg. Area Income']),y=data['Price'])
data.plot(kind='scatter',x='Avg. Area Income',y='Price',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['Avg. Area Income']))
plt.plot(data['Avg. Area Income'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Avg. Area Income'],data['Avg. Area Income']**2,data['Avg. Area Income']**3,data['Avg. Area Income']**4,data['Avg. Area Income']**5]).T
model.fit(X,y=data['Price'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['Price']))
import numpy as np
range = np.arange(0,3,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred=model.predict(X_range)
data.plot(kind='scatter',x='Avg. Area Income',y='Price',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
y='Price'
model.fit(X=data.loc[:,['Avg. Area Income','Avg. Area House Age']],y=data['Price'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X=data.loc[:,['Avg. Area Income','Avg. Area House Age']],y=data['Price']))
```

```python
model=linear_model.LinearRegression()
y='Price'
model.fit(X=data.loc[:,['Avg. Area Income','Avg. Area House Age','Area Population']],y=data['Price'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X=data.loc[:,['Avg. Area Income','Avg. Area House Age','Area Population']],y=data['Price']))
```

```python
model=linear_model.LinearRegression()
y='Price'
model.fit(X=data.loc[:,['Avg. Area Income','Avg. Area House Age','Area Population','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms']],y=data['Price'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X=data.loc[:,['Avg. Area Income','Avg. Area House Age','Area Population','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms']],y=data['Price']))
```
